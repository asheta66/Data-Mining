```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original data as a 10x2 numpy array
X = np.array([
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
])

# Compute the mean for each feature
mean = np.mean(X, axis=0)  # mean = [1.81, 2.01]

# Zero-center the data
X_centered = X - mean

# Plotting
plt.figure(figsize=(9, 5))

# Original Data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='blue', label='Original Data')
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.legend()

# Zero-Centered Data
plt.subplot(1, 2, 2)
plt.scatter(X_centered[:, 0], X_centered[:, 1], color='red', label='Zero-Centered Data')
plt.title('Zero-Centered Data')
plt.xlabel('Feature 1 (centered)')
plt.ylabel('Feature 2 (centered)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
